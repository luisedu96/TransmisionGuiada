```
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Math, Latex

uo=1.25566e-6
eo=8.8542e-12
c=3e8

f=1e3
T=1/f
er=1
ur=1
sigma=5.8e7

e=eo*er
u=uo*ur

omega=2*np.pi*f
tand=sigma/(omega*e)

alpha = omega*np.sqrt(u*e/2)*np.sqrt(np.sqrt(1+tand**2)-1)
beta  = omega*np.sqrt(u*e/2)*np.sqrt(np.sqrt(1+tand**2)+1)
delta = 1/alpha
lamb = 2*np.pi/beta

z = np.arange(0, 0.015, 1e-5)
envol = np.exp(-alpha*z)
Ez = envol*np.cos(-beta*z)

f, ax = plt.subplots(1, figsize=(10, 7))
rec_r = plt.Rectangle((0, -1.2), delta, 2.4, color='orange',alpha=0.3)
rec_b = plt.Rectangle((delta, -1.2), z[len(z)-1], 2.4, color='b',alpha=0.1)
ax.add_artist(rec_r)
ax.add_artist(rec_b)
ax.plot(z, envol,'k--',label='Envolvente',lw=1.5)
ax.plot(z, Ez, label='E',lw=3)
ax.plot([0, 0.015], [0.37, 0.37], 'r--', lw=1.5, label='0.37')
ax.axis([0, 0.015, -1.2, 1.2])
ax.set_xlabel('Distancia Z [m]', fontsize=14)
ax.set_ylabel('E (V/m)', fontsize=14)
ax.legend()
ax.set_title('Campo eléctrico propagándose'+'\n'+' en un material conductor, $\sigma$='+str(sigma), fontsize=14)
plt.show()
```
